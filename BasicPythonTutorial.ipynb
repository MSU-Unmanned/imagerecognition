### Nesting Functions
You can also nest functions within each other or have combinations of functions. 

```python
#nesting functions, can be nested in any order as long as the output of one is the same form as the input as another
print('Nesting numpy functions:', np.add(np.subtract(100, 17), np.multiply(4, 5)))

#python will follow order of operatins by default, but you can also add parentheses to stay safe
print('No paretheses order of operations:', 2*3 + 6/3)
```

It's important to pay attention to order of operations and what you expect a function to do, especially when nesting functions.

Here are some examples:
1. 3 + 5 - 8 / 2^2 = 6
2. 3 + 5 - (8 / 2)^2 = -8
3. (3 + 5 - 8) / 2^2 = 0
4. 3 + ((5 - 8) / 2)^2 = 21/4

Below are a few examples of the order of operations possible to get the right answers for each.

In general, the fewer operations we perform, the better, because it emliminates the number of chances we have to make mistakes and speeds up our code.

It's also always important to make sure you know what you expect from your code. In these examples, I worked them out then coded each on up. In the 4th example I initially left out a set of parentheses and got the wrong answer, so use caution with parentheses and order of operations.

```python
print('Example 1:')
print('Built-in:', 3 + 5 - 8 / 2**2)
print('Numpy:', np.subtract(np.add(3,5), np.divide(8, np.power(2,2))))
print('Numpy:', np.add(3, np.subtract(5, np.divide(8, np.power(2,2)))))


print('Example 2:')
print('Built-in:', 3 + 5 - (8 / 2)**2)
print('Numpy:', np.subtract(np.add(3,5), np.power(np.divide(8,2),2)))
print('Numpy:', np.add(3, np.subtract(5, np.power(np.divide(8,2),2))))


print('Example 3:')
print('Built-in:', (3 + 5 - 8) / 2**2)
print('Numpy:', np.divide(np.subtract(np.add(3,5), 8), np.power(2,2)))
print('Numpy:', np.divide(np.add(3, np.subtract(5, 8)), np.power(2,2)))


print('Example 4:')
print('Built-in:', 3 + ((5 - 8) / 2)**2)
print('Numpy:', np.add(3, np.power(np.divide(np.subtract(5, 8), 2),2)))
print('Numpy:', np.add(3, np.divide(np.power(np.subtract(5, 8),2), np.power(2,2))))
```

### Sympy

Sympy helps us print out pretty results for out matricies and get "row reduced echelon form" (rref) from our matricies.

RREF will help us solve sets of equations and easily see the output. It's a linear algebra concept that can be done by hand, but for our purposes we will not. .rref() will work on matricies of any size/shape, so even if it's not a square matrix we can get our solutin through the code.

Note: sympy's matrix input format is the same as numpy's, but sympy's functions is sym.Matrix() with a capital M, unlike numpy's lowercase np.matrix(). Python is a case sensitive language, so mixing up the capitalization will lead to an error in your code.

```python
import sympy as sym

# np.Matrix([[1]]) #Uncomment this to see the error
print(np.matrix([[1]]))

# sym.matrix([[1]]) #Uncomment this to see the error
sym.Matrix([[1]])
```

When creating matricies, you need to use two layers of brackets.
Ex:
1. Good: [[1,2], [3,4]]
2. Bad: [1,2], [3,4]

Small warning: For both numpy and sympy, they will make matrices where at least one dimension is 1 even without the proper formatting.
* Both will make a matrix with a list as an input
    * Ex: [x1, x2, ..., xn]
    * numpy will make the list into a 1xn matrix where it has one row and n columns.
    * sympy will make the list into an nx1 matrix where it has one column and n rows.
    * if n=1, both will make the input into a 1x1 matrix
* numpy will make a 1x1 matrix with just an integer input

When printing out the matrices, python has no special formatting, so using print() will show the same output as having it be the last item in the cell. Sympy on the other hand has special formatting, so we need to use display() instead of print() to maintain the formatting.

```python
#Wrongly formatted matrices
print(np.matrix(1))
print(np.matrix([1,2]))
# np.matrix(1,2)

# print(sym.Matrix(1))
display(sym.Matrix([1]))
sym.Matrix([1,2,3])
```

```python
#Correctly formatted matrices -> Use two sets of brackets
print(np.matrix([[1,2], [3,4]])) #2x2
print(np.matrix([[1,2], [3,4], [5,6], [7,8], [9,0], [132, 109]])) #6x2
print(np.matrix([[1,2,3,4,5,6], [7,8,9,0,132,109]])) #2x6


display(sym.Matrix([[1,2], [3,4]])) #2x2
display(sym.Matrix([[1,2], [3,4], [5,6], [7,8], [9,0], [132, 109]])) #6x2
sym.Matrix([[1,2,3,4,5,6], [7,8,9,0,132,109]]) #2x6
```

### Examples

One example we will use below is the matrix [[1,2,3],[4,5,6]]. There are a few ways we can organize systems of equations.

1. Row form
$$\begin{align}
1x + 2y &= 3\\
4x + 5y &= 6
\end{align}$$

2. Column form
$$x\begin{bmatrix}1 \\ 4\end{bmatrix} +
  y\begin{bmatrix}2 \\ 5\end{bmatrix} = \begin{bmatrix}3 \\ 6\end{bmatrix}$$

3. Augmented matrix form
$$\left[ \begin{array}{rr|r}
1 & 2 & 3  \\
4 & 5 & 6 
\end{array}
\right]$$

3. Ax=b
$$\begin{bmatrix}1 & 2 \\ 4 & 5\end{bmatrix} \begin{bmatrix}x \\ y\end{bmatrix}
   = \begin{bmatrix}3 \\ 6\end{bmatrix}$$

```python
print(sym.Matrix([[1,2,3], [4,5,6]]).rref()[0]) #Using the print function will remove the formating
display(sym.Matrix([[1,2,3], [4,5,6]]).rref()[0]) #Using the display function will not
sym.Matrix([[1,2,3], [4,5,6]]).rref()[0]
```

The output for sympy's rref() function is a tuple (you don't need to understand what that it, it's just a python data type), which includes the RREF matrix, which is the output we want, and another tuple, an output we don't particularly care about. So we displaye the rref matrix by using .rref()[0] to select just the matrix.

Below is an example of the whole output. Above is an example of sympy's formatting

```python
sym.Matrix([[1,2,3], [4,5,6]]).rref()
```

```python
M = np.matrix([[1,2,3], [4,5,6]])
M #numpy has no special display formatting for their matricies
```

If we use the form Ax=b, we can use numpy to solve for the x array. Numpy's function to solve this does require a square matrix though.

```python
A = np.matrix([[1,2], [4,5]])
b = np.matrix([[3], [6]])
x = np.linalg.solve(A,b)
x #These values are the same we just saw in the rref, so we're all set!
```

```python
from tensorflow.keras import losses
from tensorflow.keras import datasets, layers, models
import cv2
from tensorflow.keras.preprocessing.sequence import pad_sequences

from PIL import Image
# from sklearn.model_selection import train_test_split


#We already have these imported above, but if you try to copy this code, you'll need to uncomment them
# import matplotlib.pyplot as plt
# import numpy as np
```

```python
#Loads data
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()
train_images, test_images = train_images / 255.0, test_images / 255.0

#Labels for pictures in words instead of unordered numbers
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']


#Plotting the images with labels
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i])
    # The CIFAR labels happen to be arrays, which is why you need the extra index
    plt.xlabel(class_names[train_labels[i][0]]) #for database data

plt.show() #This will show some of the images
```

```python
#Initializing the model
model = models.Sequential()

#Adding layers to the model
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))

#Adding more layers to the model
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10))

#Printing the model's summary
model.summary()
```
